```python
import subprocess
from IPython.display import Audio, display

def generate_and_play_audio(text, voice_name, label):
    command = ['edge-tts', '--voice', voice_name, '--text', text, '--write-media', 'edge.mp3']
    subprocess.run(command, check=True, stdout=subprocess.DEVNULL)
    print(f"Playing audio for {label} ({voice_name}):")
    display(Audio("edge.mp3", autoplay=False))

# Indonesian examples
indonesian_text = "Halo, nama saya Ardi. Selamat datang di Azure Text-to-Speech."
indonesian_voices = [
    ("id-ID-ArdiNeural", "Ardi"),
    ("id-ID-GadisNeural", "Gadis")
]

print("Indonesian Examples:")
for voice_name, label in indonesian_voices:
    generate_and_play_audio(indonesian_text, voice_name, label)

# English examples
english_text = "Hello, my name is Andrew. Welcome to Azure Text-to-Speech."
english_voices = [
    ("en-AU-NatashaNeural", "Natasha (Australia)"),
    ("en-AU-WilliamNeural", "William (Australia)"),
    ("en-CA-ClaraNeural", "Clara (Canada)"),
    ("en-CA-LiamNeural", "Liam (Canada)"),
    ("en-GB-LibbyNeural", "Libby (UK)"),
    ("en-GB-MaisieNeural", "Maisie (UK)"),
    ("en-GB-RyanNeural", "Ryan (UK)"),
    ("en-GB-SoniaNeural", "Sonia (UK)"),
    ("en-GB-ThomasNeural", "Thomas (UK)"),
    ("en-HK-SamNeural", "Sam (Hong Kong)"),
    ("en-HK-YanNeural", "Yan (Hong Kong)"),
    ("en-IE-ConnorNeural", "Connor (Ireland)"),
    ("en-IE-EmilyNeural", "Emily (Ireland)"),
    ("en-IN-NeerjaExpressiveNeural", "Neerja Expressive (India)"),
    ("en-IN-NeerjaNeural", "Neerja (India)"),
    ("en-IN-PrabhatNeural", "Prabhat (India)"),
    ("en-KE-AsiliaNeural", "Asilia (Kenya)"),
    ("en-KE-ChilembaNeural", "Chilemba (Kenya)"),
    ("en-NG-AbeoNeural", "Abeo (Nigeria)"),
    ("en-NG-EzinneNeural", "Ezinne (Nigeria)"),
    ("en-NZ-MitchellNeural", "Mitchell (New Zealand)"),
    ("en-NZ-MollyNeural", "Molly (New Zealand)"),
    ("en-PH-JamesNeural", "James (Philippines)"),
    ("en-PH-RosaNeural", "Rosa (Philippines)"),
    ("en-SG-LunaNeural", "Luna (Singapore)"),
    ("en-SG-WayneNeural", "Wayne (Singapore)"),
    ("en-TZ-ElimuNeural", "Elimu (Tanzania)"),
    ("en-TZ-ImaniNeural", "Imani (Tanzania)"),
    ("en-US-AnaNeural", "Ana (US)"),
    ("en-US-AndrewMultilingualNeural", "Andrew Multilingual (US)"),
    ("en-US-AndrewNeural", "Andrew (US)"),
    ("en-US-AriaNeural", "Aria (US)"),
    ("en-US-AvaMultilingualNeural", "Ava Multilingual (US)"),
    ("en-US-AvaNeural", "Ava (US)"),
    ("en-US-BrianMultilingualNeural", "Brian Multilingual (US)"),
    ("en-US-BrianNeural", "Brian (US)"),
    ("en-US-ChristopherNeural", "Christopher (US)"),
    ("en-US-EmmaMultilingualNeural", "Emma Multilingual (US)"),
    ("en-US-EmmaNeural", "Emma (US)"),
    ("en-US-EricNeural", "Eric (US)"),
    ("en-US-GuyNeural", "Guy (US)"),
    ("en-US-JennyNeural", "Jenny (US)"),
    ("en-US-MichelleNeural", "Michelle (US)"),
    ("en-US-RogerNeural", "Roger (US)"),
    ("en-US-SteffanNeural", "Steffan (US)"),
    ("en-ZA-LeahNeural", "Leah (South Africa)"),
    ("en-ZA-LukeNeural", "Luke (South Africa)")
]

print("English Examples:")
for voice_name, label in english_voices:
    generate_and_play_audio(english_text, voice_name, label)
```
